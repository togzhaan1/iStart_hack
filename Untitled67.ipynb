# Car Scratch and Dent Detection with YOLOv8

This notebook demonstrates how to train a YOLOv8 model to detect car scratches and dents using a custom dataset from Roboflow.

## Setup

1.  **Install Dependencies**: The notebook starts by installing the necessary libraries, `ultralytics` and `roboflow`.
2.  **Download Dataset**: The dataset is downloaded from Roboflow using your API key. Make sure to replace the placeholder API key with your actual key and update the workspace and project details if necessary.

## Training

The YOLOv8 model is trained on the downloaded dataset for a specified number of epochs. The training process outputs various metrics and saves the trained model weights.

## Prediction

After training, the best trained model is loaded and used to make predictions on a sample image. The detected scratches and dents are then visualized on the image.

## How to Use

1.  Run all the cells in the notebook sequentially.
2.  Replace the placeholder Roboflow API key and project details with your own.
3.  Update the `source` path in the prediction cell to point to the image you want to test.
4.  Adjust the training parameters (epochs, batch size, etc.) as needed for your specific dataset and requirements.

## Results

The trained model will produce bounding box predictions around detected car scratches and dents in the input image. The results are saved to a directory within the notebook environment.
